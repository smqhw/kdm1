```
from spacy.lemmatizer import Lemmatizer
from spacy.lookups import Lookups
sent=nlp("Xi Jinping is a Chinese politician who has served as General Secretary of the Chinese Communist Party") 
for word in sent:
 print(word.text,"=>",word.lemma_)
```

```
from nltk import word_tokenize, pos_tag
text = "Xi Jinping is a Chinese politician who has served as General Secretary of the Chinese Communist Party"
tokens =nltk.word_tokenize(text)
pos_tag(tokens)
```

```
from nltk import ne_chunk
NE_sent="Xi Jinping is a Chinese politician who has served as General Secretary of the Chinese Communist Party "
NE_tokens=word_tokenize(NE_sent)
NE_tags=nltk.pos_tag(NE_tokens)
NE_NER=ne_chunk(NE_tags)
print(NE_NER)
```

```
new="Xi Jinping is a Chinese politician who has served as General Secretary of the Chinese Communist Party"
new_tokens=nltk.pos_tag(word_tokenize(new))
new_tokens
grammar_np= r"NP:{<DT>?<JJ>*<NN>}"
chunk_parser=nltk.RegexpParser(grammar_np)
chunk_result=chunk_parser.parse(new_tokens)
chunk_result
```
